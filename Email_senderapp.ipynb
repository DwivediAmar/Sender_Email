```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders
import tkinter as tk
from tkinter import messagebox, filedialog

def send_email(from_email, from_password, subject, body, to_email, attachment_path=None):
    smtp_server = "smtp.gmail.com"
    smtp_port = 587

    # Create the MIME message
    msg = MIMEMultipart()
    msg['From'] = from_email
    msg['To'] = to_email
    msg['Subject'] = subject

    # Attach the body with the msg instance
    msg.attach(MIMEText(body, 'plain'))

    # Attach file if specified
    if attachment_path:
        try:
            with open(attachment_path, "rb") as attachment:
                part = MIMEBase('application', 'octet-stream')
                part.set_payload(attachment.read())
                encoders.encode_base64(part)
                part.add_header('Content-Disposition', f'attachment; filename={attachment_path.split("/")[-1]}')
                msg.attach(part)
        except Exception as e:
            print(f"Error attaching file: {e}")
            messagebox.showerror("Error", f"Error attaching file: {e}")
            return

    try:
        # Set up the SMTP server and send the email
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls()  # Secure the connection
        server.login(from_email, from_password)  # Use the app password here if 2FA is enabled
        text = msg.as_string()
        server.sendmail(from_email, to_email, text)
        server.quit()
        print(f"Email sent successfully to {to_email}")
    except Exception as e:
        print(f"Failed to send email: {str(e)}")
        messagebox.showerror("Error", f"Failed to send email: {str(e)}")


# Function to handle the button click event
def on_send_email_button_click():
    from_email = from_email_entry.get()
    from_password = from_password_entry.get()
    subject = subject_entry.get()
    body = body_entry.get("1.0", tk.END)  # Get text from multiline entry
    to_email = to_email_entry.get()

    # Get the attachment file path
    attachment_path = attachment_entry.get()

    # Call the send_email function
    try:
        send_email(from_email, from_password, subject, body, to_email, attachment_path if attachment_path else None)
        messagebox.showinfo("Success", "Email sent successfully!")
    except Exception as e:
        messagebox.showerror("Error", f"Failed to send email: {str(e)}")

# Function to open a file dialog for selecting the file
def browse_file():
    filename = filedialog.askopenfilename(title="Select a file to attach", filetypes=[("All files", "*.*")])
    attachment_entry.delete(0, tk.END)  # Clear the existing entry
    attachment_entry.insert(0, filename)  # Set the selected file path

# Create the main window
root = tk.Tk()
root.title("Send Email with Attachment")

# Set the window size (larger frame)
root.geometry("600x600")  # Adjust this size to your needs

# Set a large font size for all widgets
font_style = ('Arial', 14)

# Create and place the labels and entry fields for the email details

# Sender Email
tk.Label(root, text="Sender Email:", font=font_style).grid(row=0, column=0, padx=20, pady=10, sticky="w")
from_email_entry = tk.Entry(root, width=40, font=font_style)
from_email_entry.grid(row=0, column=1, padx=20, pady=10)

# Sender Password
tk.Label(root, text="Sender Password:", font=font_style).grid(row=1, column=0, padx=20, pady=10, sticky="w")
from_password_entry = tk.Entry(root, show="*", width=40, font=font_style)
from_password_entry.grid(row=1, column=1, padx=20, pady=10)

# Recipient Email
tk.Label(root, text="To Email:", font=font_style).grid(row=2, column=0, padx=20, pady=10, sticky="w")
to_email_entry = tk.Entry(root, width=40, font=font_style)
to_email_entry.grid(row=2, column=1, padx=20, pady=10)

# Subject
tk.Label(root, text="Subject:", font=font_style).grid(row=3, column=0, padx=20, pady=10, sticky="w")
subject_entry = tk.Entry(root, width=40, font=font_style)
subject_entry.grid(row=3, column=1, padx=20, pady=10)

# Body
tk.Label(root, text="Body:", font=font_style).grid(row=4, column=0, padx=20, pady=10, sticky="w")
body_entry = tk.Text(root, width=40, height=10, font=font_style)
body_entry.grid(row=4, column=1, padx=20, pady=10)

# Attachment Section
tk.Label(root, text="Attachment:", font=font_style).grid(row=5, column=0, padx=20, pady=10, sticky="w")
attachment_entry = tk.Entry(root, width=40, font=font_style)
attachment_entry.grid(row=5, column=1, padx=20, pady=10)

# Browse Button
browse_button = tk.Button(root, text="Browse", font=font_style, command=browse_file)
browse_button.grid(row=5, column=2, padx=20, pady=10)

# Create the Send button with color
send_button = tk.Button(root, text="Send", font=font_style, command=on_send_email_button_click, bg="lightblue", fg="white")
send_button.grid(row=6, column=0, columnspan=3, pady=20)


# Run the GUI loop
root.mainloop()
```
